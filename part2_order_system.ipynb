```python
# ============================================================
# TASK 1 - Explore the Menu
# ============================================================

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}
# Let's see what's inside one item
print(menu["Paneer Tikka"])
print(menu["Paneer Tikka"]["price"])
print(menu["Paneer Tikka"]["available"])
# Define categories in order
categories = ["Starters", "Mains", "Desserts"]

# Loop through each category
for category in categories:
    print(f"\n===== {category} =====")

    # Loop through every item in menu
    for item_name, item_info in menu.items():

        # Only print items that belong to this category
        if item_info["category"] == category:

            # Check availability
            if item_info["available"] == True:
                status = "[Available]"
            else:
                status = "[Unavailable]"

            # Print formatted row
            print(f"{item_name:<16} ₹{item_info['price']:<8.2f} {status}")
            # Total number of items
total_items = len(menu)
print(f"\nTotal items on menu : {total_items}")

# Total available items
available_items = sum(1 for item in menu.values() if item["available"] == True)
print(f"Total available items : {available_items}")

# Most expensive item
most_expensive = max(menu, key=lambda x: menu[x]["price"])
print(f"Most expensive item : {most_expensive} ₹{menu[most_expensive]['price']}")

# Items priced under ₹150
print("\nItems under ₹150:")
for item_name, item_info in menu.items():
    if item_info["price"] < 150:
        print(f"  {item_name} : ₹{item_info['price']}")
```

```python
# ============================================================
# TASK 2 - Cart Operations
# ============================================================

cart = []

# --- Function to display cart ---
def show_cart(cart):
    print("\nCurrent Cart:")
    if len(cart) == 0:
        print("  Cart is empty!")
    else:
        for entry in cart:
            print(f"  {entry['item']} x{entry['quantity']} ₹{entry['price'] * entry['quantity']:.2f}")

# --- Logic to add item to cart ---
def add_to_cart(cart, item_name, quantity):
    # Check if item exists in menu
    if item_name not in menu:
        print(f"❌ '{item_name}' does not exist in the menu!")
        return

    # Check if item is available
    if menu[item_name]["available"] == False:
        print(f"❌ '{item_name}' is currently unavailable!")
        return

    # Check if item already exists in cart
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f"✅ Updated '{item_name}' quantity to {entry['quantity']}")
            return

    # Item not in cart yet - add new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"✅ Added '{item_name}' x{quantity} to cart!")

# --- Test ---
add_to_cart(cart, "Paneer Tikka", 2)
show_cart(cart)
# Add Gulab Jamun x1
add_to_cart(cart, "Gulab Jamun", 1)
show_cart(cart)

# Add Paneer Tikka x1 again - should update quantity to 3
add_to_cart(cart, "Paneer Tikka", 1)
show_cart(cart)

# Try to add item that doesn't exist
add_to_cart(cart, "Mystery Burger", 1)
show_cart(cart)

# Try to add unavailable item
add_to_cart(cart, "Chicken Wings", 1)
show_cart(cart)
# --- Logic to remove item from cart ---
def remove_from_cart(cart, item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"✅ Removed '{item_name}' from cart!")
            return
    print(f"❌ '{item_name}' is not in the cart!")

# Remove Gulab Jamun
remove_from_cart(cart, "Gulab Jamun")
show_cart(cart)
# --- Order Summary ---
print("\n========== Order Summary ==========")

subtotal = 0
for entry in cart:
    item_total = entry["price"] * entry["quantity"]
    subtotal += item_total
    print(f"{entry['item']:<20} x{entry['quantity']}    ₹{item_total:.2f}")

gst = round(subtotal * 0.05, 2)
total_payable = round(subtotal + gst, 2)

print("------------------------------------")
print(f"{'Subtotal:':<25}  ₹{subtotal:.2f}")
print(f"{'GST (5%):':<25}  ₹{gst:.2f}")
print(f"{'Total Payable:':<25}  ₹{total_payable:.2f}")
print("====================================")
```

```python
# ============================================================
# TASK 3 - Inventory Tracker with Deep Copy
# ============================================================

import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# Deep copy inventory into backup
inventory_backup = copy.deepcopy(inventory)

# Demonstrate deep copy works - change one value in inventory
inventory["Paneer Tikka"]["stock"] = 999
print("After change:")
print(f"  inventory stock       : {inventory['Paneer Tikka']['stock']}")
print(f"  inventory_backup stock: {inventory_backup['Paneer Tikka']['stock']}")

# Restore inventory back to original
inventory["Paneer Tikka"]["stock"] = 10
print("\nInventory restored back to original!")
# Deduct cart items from inventory
print("\nProcessing order fulfilment...")
for entry in cart:
    item_name = entry["item"]
    quantity = entry["quantity"]
    current_stock = inventory[item_name]["stock"]

    if current_stock >= quantity:
        inventory[item_name]["stock"] -= quantity
        print(f"✅ Deducted {quantity} of '{item_name}'. Stock left: {inventory[item_name]['stock']}")
    else:
        print(f"⚠️ Insufficient stock for '{item_name}'! Only {current_stock} left. Deducting what's available.")
        inventory[item_name]["stock"] = 0
        # --- Reorder Alert ---
print("\nChecking reorder levels...")
for item_name, item_info in inventory.items():
    if item_info["stock"] <= item_info["reorder_level"]:
        print(f"⚠️ Reorder Alert: {item_name} — Only {item_info['stock']} unit(s) left (reorder level: {item_info['reorder_level']})")
        # Temporary test - manually set Rasgulla stock low
inventory["Rasgulla"]["stock"] = 2
print("Testing reorder alert:")
for item_name, item_info in inventory.items():
    if item_info["stock"] <= item_info["reorder_level"]:
        print(f"⚠️ Reorder Alert: {item_name} — Only {item_info['stock']} unit(s) left (reorder level: {item_info['reorder_level']})")

# Restore back
inventory["Rasgulla"]["stock"] = 4
print("Rasgulla stock restored!")
# --- Print both inventory and backup to prove they differ ---
print("\n--- Current Inventory (after deduction) ---")
for item_name, item_info in inventory.items():
    print(f"  {item_name:<16} stock: {item_info['stock']}")

print("\n--- Inventory Backup (original - unchanged) ---")
for item_name, item_info in inventory_backup.items():
    print(f"  {item_name:<16} stock: {item_info['stock']}")

print("\n✅ Deep copy proved! Backup still shows original stock values.")
```

```python
# ============================================================
# TASK 4 - Daily Sales Log Analysis
# ============================================================

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# --- Total revenue per day ---
print("Revenue Per Day:")
print("-" * 30)
for date, orders in sales_log.items():
    daily_total = sum(order["total"] for order in orders)
    print(f"  {date} : ₹{daily_total:.2f}")

    # --- Best selling day ---
best_day = max(sales_log, key=lambda date: sum(order["total"] for order in sales_log[date]))
best_day_total = sum(order["total"] for order in sales_log[best_day])
print(f"\nBest Selling Day : {best_day} with ₹{best_day_total:.2f}")

# --- Most ordered item ---
item_count = {}
for date, orders in sales_log.items():
    for order in orders:
        for item in order["items"]:
            if item in item_count:
                item_count[item] += 1
            else:
                item_count[item] = 1

most_ordered = max(item_count, key=lambda x: item_count[x])
print(f"Most Ordered Item : {most_ordered} ({item_count[most_ordered]} orders)")
# --- Add new day to sales_log ---
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

# Reprint revenue per day after adding new day
print("\nUpdated Revenue Per Day:")
print("-" * 30)
best_day_total = 0
best_day = ""
for date, orders in sales_log.items():
    daily_total = sum(order["total"] for order in orders)
    if daily_total > best_day_total:
        best_day_total = daily_total
        best_day = date
    print(f"  {date} : ₹{daily_total:.2f}")

print(f"\nUpdated Best Selling Day : {best_day} with ₹{best_day_total:.2f}")

# --- Numbered list of all orders ---
print("\nAll Orders:")
print("-" * 60)
counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"{counter}.  [{date}] Order #{order['order_id']}  — ₹{order['total']:.2f} — Items: {items_str}")
        counter += 1
```
